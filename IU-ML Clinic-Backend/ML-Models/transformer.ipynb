```python
import tf2onnx
import tensorflow as tf

# Hata almamak için compile=False kullanmaya devam ediyoruz
model = tf.keras.models.load_model('./ResNet50_best_tumor_classifier.h5', compile=False)

# Modeli ONNX formatına çevir
# input_signature: Modelin girdi şeklini belirtir.
# opset: ONNX versiyonu, 13 genellikle uyumludur.
spec = (tf.TensorSpec((None, 128, 128, 3), tf.float32, name="input"),)
output_path = "ResNet50_best_tumor_classifier.onnx"

model_proto, _ = tf2onnx.convert.from_keras(model, input_signature=spec, opset=13)
with open(output_path, "wb") as f:
    f.write(model_proto.SerializeToString())

print(f"\nSUCCESS: Model successfully converted to ONNX format and saved as '{output_path}'")
```
